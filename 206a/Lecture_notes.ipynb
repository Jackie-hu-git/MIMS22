### Circular, doubly linked lists(with a sentinel)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None 
        self.prev = None
        
class Sentinel_DLL:
    def __init__(self):
        self.sentinel = Node(None)
        self.sentinel.next = self.sentinel
        self.sentinel.prev = self.sentinel 
    
    # Return a reference to the first node in the list, if there is one. 
    # If the list is empty, return None
    
    def first_node(self):
        if self.sentinel.next == self.sentinel:
            return None
        else:
            return self.sentinel.next
    
    # insert the Node, with data, x is the Node we want to insert after 
    def insert_after(self, x, data):
        y = Node(data) # make a new Node object
        z = x.next # y goes between x and z 
        
        y.prev = x
        y.next = z
        x.next = y
        z.prev = y
        
    # Insert a new node at the end of the list
    def append(self, data):
        last_node = self.sentinel.prev # find the end the of the Node in the list 
        self.insert_after(last_node, data) # use the previous insert func for append 
    
    # Delete node x from the list 
    def delete(self, x):
        x.prev.next = x.next 
        x.next.prev = x.prev 
    
    
    # find a node containing data, and return a reference to it
    def find(self, data):
        self.sentinel.data = data # store copy of data in the sentinel
        
        x = self.first_node()
        
        while x.data != data:
            x = x.next
                
        '''
        either find the data in the list
        or looped around find it in the sentinel
        
        '''
        self.sentinel.data = None # Restore the sentinel's data 
        
        if x == self.sentinel:
            return None #data wasn't really in the list
        else:
            return x # we find it in the list 
        
        # Return kth node in the list
        def random_access(self, k):
            counter = 0
            x = self.first_node()
            
            while counter < k and x != sentinel: # not loop aounrd 
                x = x.next 
                counter += 1
                
            if x == self.sentinel:
                return None # list has fewer than k elements
            else: 
                return x # return kth node 
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None 
        self.next = None 
        
    def __str__(self):
        
        return str(self.data)
        
#Dll without sentinel     
class DLL:
    def __init__(self):
        
        self.first = Node('cat')
        self.first.next = self.first
        self.first.prev = self.first 
    
    #add a node to the DLL, given it's data   
    def add_to_last(self, d):
        last_node = self.first.prev
        new_node = Node(d)
        
        last_node.next = new_node
        new_node.prev = last_node
        new_node.next = self.first 
        self.first.prev = new_node
    

    def __str__(self):
        cur = self.first
        s = ''
        
        while cur.next != self.first:
            s = s + cur.data
            cur = cur.next
        
        s += cur.data
        
        return s 
```

### Dictionary

```python
#Create a word histogram for DOI
def printSortedDictionary(D):
    DS = sorted(D.items(), key=lambda x:x[1])
    for k in DS:
        print(k)
```

```python
#strip word of punctuation and convert to all lower0case
def stripWork(w):
    w = w.replace( ".", "" )
    w = w.replace( ",", "" )
    w = w.replace( ";", "" )
    w = w.replace( ":", "" )
    w = w.replace( "'", "" )
    w = w.replace( "&", "" )
    w = w.replace( "\n", "" )
    w = w.lower()
    return( w )
    
```

```python
myfile = open( 'pk1.txt', 'r',encoding="utf8")
doi = {}
N = 0
# go through each lines, seperated by space 
for line in myfile:
    line = line.split(' ')
    #go through each words in each line, and convert it to lowercase and get rid of the punc.
    for word in line:
        word = stripWork(word)
        # make sure it's a word, and tally
        if(len(word) > 0):
            if word in doi:
                doi[word] += 1
                N += 1
            else:
                doi[word] = 1
                N += 1
myfile.close()
printSortedDictionary(doi)
print(N)
```

### Graphs

```python
class Node:
    def __init__(self, data):
        self.label = data
        self.left = None
        self.right = None
        
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
```

```python
# preorder traversal: root, left sibtree, right subtree 
# 1 | 2 | 4 5 | 3 | 6

def traversePreOrder(self):
    print(self.label, end= ' ')
    if self.left:
        self.left.traversePreOrder()
    if self.right:
        self.right.traversePreOrder()
```

```python
# inorder: left subtree, root, right subtree
def tracerseInOrder(self):
    if self.left:
        self.left.traverseInOrder()
    print(self.label, end= ' ')
    if self.right:
        self.right.traverseInOrder()
```

```python
#post order: left subtree, right subtree, root
def traversePostOrder(self):
    if self.left:
        self.left.traversePostOrder()
    if self.right:
        self.right.traverseInOrder()
    print(self.label, end= ' ')
```

```python
def b(A):
    n = len(A)
    l = [i for i in range(n)]
    s = True
    while s:
        s = False
        for i in range (0, n -1):
            if A[i] > A[i+1]:
                temp = A[i]
                A[i] = A[i + 1]
                A[i + 1] = temp
                
                t = l[i]
                l[i] = l[i + 1]
                l[i + 1] = t
                
                s = True
    return l
```
