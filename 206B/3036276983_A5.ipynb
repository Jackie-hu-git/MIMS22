#### build a dictionary consisting of each unique word in a document along with the frequency with which that word is used in the document.


```python
# convert a word to lower-case and remove all punctuation
def strip_word( w ):
    w = w.lower()
    w = ''.join([i for i in w if i.isalpha()])
    return( w )
```

```python
#Sort the dictionary in order of most to least most frequent words
def sort_dictionary( D ):
    Ds = sorted(D.items(), key=lambda x:x[1], reverse=True)               
    return Ds
```

```python
#a function print_dictionary that prints each word and associated frequency count. 
def printSortedDictionary(D):
    DS = sorted(D.items(), key=lambda x:x[1])
    for k in DS:
        print(k)
```

```python
#takes an input and builds a word-count dictionary for all filenames that begin with this prefix. 
def build_dictionary(s):
    #initiate variables
    myfile_list = []
    dic = {}
    N = 0

    if s == 'pk':
    #iterate through all 9 of pk 
        for i in range(1, 10):
            myfile_list.append('pk' + str(i)+ '.txt')
    
    elif s == '':
    #iterate through all files 
       for i in range(1, 10):
            myfile_list.append('pk' + str(i) + '.txt') 
            myfile_list.append('dr' + str(i) + '.txt')

    elif s =='dr':
    #iterate through all 9 of dr files
       for i in range(1, 10):
            myfile_list.append('dr' + str(i) + '.txt') 
    
    else:
        myfile_list.append(str(s) + '.txt')
    
    
    #iterate through the file list
    for i in myfile_list:
        myfile = open(i,'r',encoding="utf8")
    
        #parse line and word
        for line in myfile:
            line = line.split(' ')

            for word in line:
                word = strip_word(word)
                
                #add count to the dictionary and N
                if(len(word) > 0):
                    if word in dic:
                        dic[word] += 1
                        N += 1
                    else:
                        dic[word] = 1
                        N += 1

    myfile.close()
    printSortedDictionary(dic)
    
    return dic,N
```

```python
#count all 18 documents 
dic,N = build_dictionary('')
for k,v in dic.items():
    v = v * (100/N)
    dic[k] = v 
    
#create a list of 25 most frequent words with it's frequency 
fr = sort_dictionary(dic)[:25]
print(fr)
```

```python
#function update the work count with dictionary with the 2nd dictionary
def buid_count(freq, dic):
    for k,v in freq.items():
        if k in dic.keys():
            freq[k] = dic[k]
        else:
            freq[k] = 0
```

```python
#build a list contains all the word count dictionaries for pk
dic_pk = []

for i in range(1, 10):
    d1,n1 = build_dictionary('pk' + str(i))
    #word count to percentag 
    for k,v in d1.items():
        v = v * (100/n1)
        d1[k] = v 
    
    dic_pk.append(d1)
```

```python
#build a list contains all the word count dictionaries for dr
dic_dr = []

for i in range(1, 10):
    d2,n2 = build_dictionary('dr' + str(i))
    #word count to percentag 
    for k,v in d2.items():
        v = v * (100/n2)
        d2[k] = v

    dic_dr.append(d2)
```

```python
#building the list of list for F1 matrix
F1_list= []
for i in range(9):
    F1_list.append(dict(fr))

#update the frequent word choice list for pk
for i in range(9):
    buid_count(F1_list[i], dic_pk[i])  
    
# each row consists of the word frequency counts for a document
F1_list_total = []

for i in F1_list:
    for v in i.values():
        F1_list_total.append(v)
```

```python
#building the list of list for F2 matrix
F2_list= []
for i in range(9):
    F2_list.append(dict(fr))

#update the frequent word choice list for dr
for i in range(9):
    buid_count(F2_list[i], dic_dr[i])
    
# each row consists of the word frequency counts for a document
F2_list_total = []

for i in F2_list:
    for v in i.values():
        F2_list_total.append(v)
```

```python
import numpy as np
F1 = np.zeros((9,25))
F2 = np.zeros((9,25))
```

```python
#sub-set list in to 9 * 25, and supply it to the matrix
t1 = np.array_split(F1_list_total, 9)
t2 = np.array_split(F2_list_total, 9)
F1 = np.matrix(t1)
F2 = np.matrix(t2)
```

```python
# Dimensionality reduction and visualization
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

F   = np.concatenate((F1,F2),axis=0)
pca = PCA(n_components=2)
Fp  = pca.fit(F).transform(F)
plt.scatter( Fp[0:9,0], Fp[0:9,1], color='b')
plt.scatter( Fp[9:18,0], Fp[9:18,1], color='r' )
```
