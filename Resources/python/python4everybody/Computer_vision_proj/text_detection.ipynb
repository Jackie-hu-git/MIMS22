```python
# adds image processing capabilities
from PIL import Image, ImageEnhance
# will convert the image to text string
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files (x86)\Tesseract-OCR\tesseract.exe"
```

```python
# assigning an image from the source path
img = Image.open("image.jpeg")
```

```python
# enhance
enhance1 = ImageEnhance.Sharpness(img)
enhance2 = ImageEnhance.Contrast(img)

img_edit = enhance1.enhance(40.0)
img_edit = enhance2.enhance(1.5)
```

```python
#save img
img_edit.save('edited.png')
# converts the image to result and saves it into result variable
result = pytesseract.image_to_string(img_edit)
```

```python
with open ("text_result.txt", mode= 'w') as file:
    file.write(result)
    print('ready')
```
